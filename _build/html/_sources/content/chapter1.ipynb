# Estructuras iterativas (ciclos)

## Instrucción `while`

- Sintaxis:  
  ```python
  while < expresión de resultado booleano > :
    < bloque de una o más instrucciones con indentado >
  ```
- Semántica:  
  - Si la `< expresión de resultado booleano >` es verdadera, ejecutar el `< bloque de una o más instrucciones con indentado >`. Una vez terminada la ejecución del bloque de instrucciones, regresar el punto de ejecución del programa a la línea donde empezó la instrucción `while`
  - Cuando la `< expresión de resultado booleano >` sea falsa, continuar con la ejecución secuencial del programa inmediatamente después del `< bloque de una o más instrucciones con indentado >`.

Nota: Una forma alternativa de interpretar la instrucción `while` es: Mientras las condición sea verdadera, volver a ejecutar el bloque de instrucciones. A diferencia del `if`, la ejecución del bloque de instrucciones se sigue repitiendo mientras la condición sea cierta.

### Ejemplo: Contador

Escriba un programa que imprima los primeros $n$ números naturales.


````
n = int( input( 'Ingrese el número n: ' ))

i = 1  # Variable que cumplirá el rol de contador

while i <= n:
  print(i)
  i = i+1  # Operación que incrementa el valor del contador en 1

print('Terminé!')
````

# Ejemplo: Acumulador

Escriba un programa que calcule la suma de los primeros $n$ números naturales.

````
n = int( input( 'Ingrese el número n: ' ))

suma = 0  # acumulador
i = 1  # contador

while i <= n:
  suma = suma + i  # acumular en suma los valores de i
  i = i+1  # incrementar el contador para producir el siguiente número natural

print('La suma es: ', suma)
````

# Asignaciones con operador

En ejemplos anteriores, para incrementar el valor de los contadores, se usó instrucciones del tipo:  
  `i = i+1`

Se puede lograr el mismo efecto usando una asignación con operador aritmético:  
  `i += 1`

En general lo operadores aritméticos: `+ - * / // % **` se pueden combinar con asignaciones. Con lo que una instrucción como:  
  `i = i*2`  
Se puede escribir también como:  
  `i *= 2`

## Ejemplo

Antes de ejecutar la siguiente celda, diga cuales son los valores que tomará `i` después de cada instrucción de asignación con operador.

````
i = 2

i += 3

i -= 1

i *= 2

i /= 8

i **= 3

i //= 3

i %= 2

print(i)
````
